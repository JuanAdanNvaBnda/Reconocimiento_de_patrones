Cargar el SCV

```python
import pandas as pd

url = 'https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv'
df = pd.read_csv(url, index_col=0) #DAta frame
```

1. ¿Cuantas variables tiene el dataframe?


```python
num_variables = df.shape[1]

num_variables
```

2. ¿Qué variables tienen valores perdidos?

```python
missing_values = df.isnull().sum()
variables_con_na = missing_values[missing_values > 0]

variables_con_na
```

3. ¿Qué variables son numéricas?

```python
variables_numericas = df.select_dtypes(include=['number']).columns

variables_numericas
```

4. ¿Qué variables son cualitativas?

```python
variables_cualitativas = df.select_dtypes(exclude=['number']).columns

variables_cualitativas
```

5. Hacer un dataframe con únicamente vinos europeos.

```python
paises_europeos = ['Albania', 'Andorra', 'Armenia', 'Austria', 'Azerbaijan', 'Belarus', 'Belgium',
                   'Bosnia and Herzegovina', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark',
                   'Estonia', 'Finland', 'France', 'Georgia', 'Germany', 'Greece', 'Hungary', 'Iceland',
                   'Ireland', 'Italy', 'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 'Moldova',
                   'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 'Poland', 'Portugal', 'Romania',
                   'Russia', 'San Marino', 'Serbia', 'Slovakia', 'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Turkey',
                   'Ukraine', 'United Kingdom', 'Vatican City']

df_europeos = df[df['country'].isin(paises_europeos)]

df_europeos
```

6. ¿Cuál es el menor, el mayor y el precio promedio de la botella por país?

```python
precios_por_pais = df_europeos.groupby('country')['price'].agg(['min', 'max', 'mean'])

precios_por_pais
```

7. ¿De que país es la botella de menor precio?

```python
pais_menor_precio = df_europeos.loc[df_europeos['price'].idxmin()]['country']

pais_menor_precio
```

8. ¿Cuantos vinos hay con *aroma a fresa* entre otras consideraciones snobs que vienen en la descripción?

```python
vinos_fresa = df[df['description'].str.contains('strawberry', case=False, na=False)]
cantidad_vinos_fresa = len(vinos_fresa)

cantidad_vinos_fresa
```

9. ¿Cuantas designaciones diferentes hay? ¿Cuál es la más repetida? ¿Cuantas veces se repite?

```python
designaciones_diferentes = df['designation'].nunique()

designaciones_diferentes
```

```python
designacion_mas_repetida = df['designation'].value_counts().idxmax()

designacion_mas_repetida
```

```python
veces_repetida = df['designation'].value_counts().max()

veces_repetida
```

10. Hacer un dataframe con la variedad, el país y el precio para vinos con un costo menor a los $20 dolares

```python
df_vinos_baratos = df[df['price'] < 20][['variety', 'country', 'price']]

df_vinos_baratos
```
